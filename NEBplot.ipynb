```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from the spline.dat file
data = np.loadtxt('spline.dat')

# Extract reaction coordinate (first column) and energy (third column)
reaction_coord = data[:, 0]
energy = data[:, 2]

# Identify the indices for labeling
initial_idx = 0  # First point
final_idx = -1   # Last point
transition_idx = np.argmax(energy)  # Highest energy point

# Calculate energy difference between initial and transition state
energy_diff = energy[transition_idx] - energy[initial_idx]

# Plot energy vs reaction coordinate
plt.figure(figsize=(8, 5))
plt.plot(reaction_coord, energy, linestyle='-', linewidth=2)

# Add labels for the initial, transition, and final points
plt.text(reaction_coord[initial_idx], energy[initial_idx], 'Initial', 
         fontsize=13, ha='center', va='bottom', color='black')
plt.text(reaction_coord[transition_idx], energy[transition_idx], 'Transition', 
         fontsize=13, ha='center', va='bottom', color='black')
plt.text(reaction_coord[final_idx], energy[final_idx], 'Final', 
         fontsize=13, ha='center', va='bottom', color='black')

# Add a vertical arrow from the initial to transition point
plt.annotate(
    '', 
    xy=(reaction_coord[transition_idx], energy[transition_idx]), 
    xytext=(reaction_coord[transition_idx], energy[initial_idx]),
    arrowprops=dict(facecolor='black', arrowstyle='<->', linewidth=2)
)

# Position the ΔE annotation to the right of the arrow
plt.text(
    reaction_coord[transition_idx] + 0.5, 
    (energy[transition_idx] + energy[initial_idx]) / 2, 
    f'ΔE = \n{energy_diff:.2f} eV', 
    fontsize=15, ha='center', va='center', color='black'
)

# Mute x-ticker
plt.xticks([])
plt.ylabel('Energy (eV)', fontsize = 15)
plt.xlabel('Reaction Coordinate', fontsize = 15)
plt.show()
```
