```python
import re
import numpy as np
import matplotlib.pyplot as plt
pattern = (
    r'\s*(\d+)\s+'                 # step
    r'T=\s*([-\d\.E+]+)\s+'        # T
    r'E=\s*([-\d\.E+]+)\s+'        # E
    r'F=\s*([-\d\.E+]+)\s+'        # F
    r'E0=\s*([-\d\.E+]+)\s+'       # E0
    r'EK=\s*([-\d\.E+]+)\s+'       # EK
    r'SP=\s*([-\d\.E+]+)\s+'       # SP
    r'SK=\s*([-\d\.E+]+)'          # SK
)
```

```python
# E0s = []
totT = []
TS = np.linspace(0, 2000, 2000)

with open('T.dat', 'r') as file:
    for line in file:
        totT.append(float(line.strip()))

# plt.plot(TS, E0s, marker='o',markersize=3)
plt.figure(figsize=(6, 4))
plt.plot(TS, totT, marker='o',markersize=3)
# plt.plot(TS, totT)
plt.xlim(0,2000)
plt.xlabel('Time (fs)', fontsize=17)
plt.ylabel('Temperature (K)', fontsize=17)
plt.tight_layout()
plt.show()
```
