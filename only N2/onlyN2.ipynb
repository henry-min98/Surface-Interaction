```python
import numpy as np
import matplotlib.pyplot as plt

def parse_xdatcar(filename):
    bond_lengths = []

    with open(filename, 'r') as file:
        lines = file.readlines()

    # Read lattice vectors
    scale = float(lines[1].strip())
    a1 = np.array([float(x) for x in lines[2].split()]) * scale
    a2 = np.array([float(x) for x in lines[3].split()]) * scale
    a3 = np.array([float(x) for x in lines[4].split()]) * scale
    lattice = np.array([a1, a2, a3]).T  # Shape (3, 3)

    # Extract atomic positions starting from line 8
    positions = []
    for line in lines:
        if line.startswith("Direct configuration="):
            i = lines.index(line)
            pos1 = np.array([float(x) for x in lines[i+1].split()])
            pos2 = np.array([float(x) for x in lines[i+2].split()])
            positions.append((pos1, pos2))

    # Compute bond lengths
    for pos1, pos2 in positions:
        cart1 = lattice @ pos1
        cart2 = lattice @ pos2
        bond_length = np.linalg.norm(cart1 - cart2)
        bond_lengths.append(bond_length)

    return bond_lengths

# --- Run and plot ---
xdatcar_file = "XDATCAR"
bond_lengths = parse_xdatcar(xdatcar_file)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(bond_lengths)+1), bond_lengths, marker='o', markersize=2)
plt.xlabel("Time Step")
plt.xlim(0,500)

plt.ylabel("N–N Bond Length (Å)")
plt.title("N₂ Bond Length vs. Time")
plt.tight_layout()
plt.show()
```
